<H1>Сепия с порогом

```python
img_copy2 = img.copy()
```

```python
depth = int(input('depth:'))
draw = ImageDraw.Draw(img_copy2) # создаем инструмент для рисования
width = img_copy2.size[0] # определяем ширину изображения
height = img_copy2.size[1] # определяем высоту изображения
pix = img_copy2.load() # выгружаем значение пикселей в память
for i in range(width):
    for j in range(height):
        r = pix[i,j][0]
        g = pix[i,j][1]
        b = pix[i,j][2]
        S = (r + g + b) // 3
        r = S + depth * 2
        g = S + depth
        b = S
        r = min(r, 255) # чтобы сумма не привысила 255
        g = min(g, 255)
        b = min(b, 255)
        draw.point((i,j), (r,g,b))
image.save("res2.jpg", "JPEG")
draw_image2(img, img_copy2)
```
